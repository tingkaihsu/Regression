```python
import numpy as np
import awkward as ak
import uproot_methods
```

```python
#This piece of code is designed for checking
#if the model recieve the correct data
##and Professor suggests that we could use mass, classifacation for later application
def SetAKArr(filepath):
    with open(filepath, 'r') as file:
        lines = file.readlines()
    #number of particle
    events = 0
    n_particles_ls = []
    px_ls = []
    py_ls = []
    pz_ls = []
    energy_ls = []
    mass_ls = []
    charge_ls = []
    #px,py,pz,engy,mass
    _label1 = []
    _label2 = []
    _label3 = []
    _label4 = []
    _label5 = []
    
    n = 0
    for line in lines:
        if line.startswith('E'):
            #particle number n is not 0
            if not n == 0:
                n_particles_ls.append(n)
                n = 0
            exp_inf = line.split()
            _label1.append(float(exp_inf[1]))
            _label2.append(float(exp_inf[2]))
            _label3.append(float(exp_inf[3]))
            _label4.append(float(exp_inf[4]))
            _label5.append(float(exp_inf[5]))
            events += 1
        else:
            par = line.split()
            ##particle +1
            n = n + 1
            px_ls.append(float(par[2]))
            py_ls.append(float(par[3]))
            pz_ls.append(float(par[4]))
            energy_ls.append(float(par[5]))
            mass_ls.append(float(par[6]))
            charge_ls.append(int(par[0]))
    if not n == 0:
        n_particles_ls.append(n)
    #change the storing type to np array
    px_arr = np.array(px_ls)
    py_arr = np.array(py_ls)
    pz_arr = np.array(pz_ls)
    energy_arr = np.array(energy_ls)
    mass_arr = np.array(mass_ls)
    charge_arr = np.array(charge_ls)
    n_particles = np.array(n_particles_ls)
    #Assign the particles into each simulation
    px = ak.JaggedArray.fromcounts(n_particles, px_arr)
    py = ak.JaggedArray.fromcounts(n_particles, py_arr)
    pz = ak.JaggedArray.fromcounts(n_particles, pz_arr)
    energy = ak.JaggedArray.fromcounts(n_particles, energy_arr)
    mass = ak.JaggedArray.fromcounts(n_particles, mass_arr)
    charge = ak.JaggedArray.fromcounts(n_particles, charge_arr)
    
    #check if total charge sum upto 1
    temp = 0
    for chgs in charge:
        for chg in chgs:
            temp += chg
        if temp != 1:
            print('Error! There is some missing charge in data!')
        temp = 0
    #check if total momentum x 
    eps = 1e-7
    temp = 0
    for i in range(0,events):
        for mx in px[i]:
            temp += mx
        if temp - _label1[i] > eps:
            print('Error! There is some missing px in data!')
        temp = 0
    #check total momentum y
    temp = 0
    for i in range(0,events):
        for my in py[i]:
            temp += my
        if temp - _label2[i] > eps:
            print('Error! There is some missing py in data!')
            print(temp)
            print(_label2[i])
        temp = 0
    #check total momentum z
    temp = 0
    for i in range(0,events):
        for mz in pz[i]:
            temp += mz
        if temp - _label3[i] > eps:
            print('Error! There is some missing pz in data!')
        temp = 0
    #check total engy
    temp = 0
    for i in range(0,events):
        for en in energy[i]:
            temp += en
        if temp - _label4[i] > eps:
            print('Error! There is some missing engy in data!')
        temp = 0
    #check total mass
    temp = 0
    for i in range(0,events):
        for ms in mass[i]:
            temp += ms
        if temp - _label5[i] > eps:
            print('Error! There is some missing py in data!')
        temp = 0
```

```python
SetAKArr('test.txt')
```
