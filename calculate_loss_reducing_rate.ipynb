```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%config InlineBackend.figure_format = 'retina'
def StoreLosses(infilepath, vaclosses = []):
    with open(infilepath, 'r') as Input:
        lines = Input.readlines()
    for line in lines:
        if line.startswith('V'):
            vac_loss = line.split()
            vaclosses.append(float(vac_loss[1]))
```

```python
def Plot(filepathes = []):
#     x = []
#     for i in range(1, len(vaclosses)+1):
#         x.append(i)
# #         print(i)
#     fig, ax = plt.subplots()
#     title = "Reduction Rate of "+str(len(vaclosses))+" epochs"
#     ax.set_title(title)
#     plt.xlabel("Epoches")
#     plt.ylabel("Loss Value")
#     ax.scatter(x,vaclosses, linewidth=2.0, color = 'red')
#     plt.legend(
#     loc='best',
#     labels = ['vac_loss', 'loss'])
# #     plt.show()
#     plt.savefig('CompareLoss.png')
    #extract every file path and add plot
    for filepath in filepathes:
        vacloss = []
        StoreLosses(filepath, vacloss)
        x = []
        label = str(len(vacloss))+' epochs'
        for i in range(1,len(vacloss)+1):
            x.append(i)
        plt.scatter(x, vacloss, label = label)
        
    # Add labels and title
    plt.xlabel('epochs')
    plt.ylabel('vac_loss')
    plt.title('Loss Decay Rate')

    # Add legend
    plt.legend()

    # Show plot
#     plt.show()
    plt.savefig('CompareLoss.png')
    
```

```python
filepathes = ["40epoches/40_loss.txt", "30epoches/30_loss.txt", "20epoches/20_loss.txt"]
Plot(filepathes)
```
